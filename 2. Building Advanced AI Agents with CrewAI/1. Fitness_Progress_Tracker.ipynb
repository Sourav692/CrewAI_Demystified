## ðŸŽ¯ Project Overview

This notebook demonstrates building an **AI-powered Fitness Progress Tracker** using CrewAI. The system consists of:

1. **Fitness Tracker Agent**: Analyzes user data and creates personalized workout plans
2. **Recommendation Agent**: Searches the web for fitness and nutrition information
3. **Coordinated Tasks**: Three tasks that work together to create a complete fitness plan

## ðŸ”‘ Key Concepts

- **Agent**: An autonomous AI entity with a specific role and goal
- **Task**: A specific assignment with clear objectives
- **Crew**: A team of agents working together
- **Tool**: External capabilities (like web search) that agents can use
- **Planning**: CrewAI's ability to intelligently coordinate task execution

## ðŸš€ Let's Get Started!


```python
# ================================================================================
# SETUP: Import Required Libraries and Initialize LLM
# ================================================================================
# This section sets up the necessary imports and initializes the Language Model
# that will power our AI agents.
# ================================================================================

from crewai import Agent, Task, Crew, LLM
from crewai_tools import SerperDevTool
import os
from dotenv import load_dotenv

# Load environment variables from .env file (contains API keys)
load_dotenv()

# Alternative: Set up API key directly (uncomment if not using .env file)
# os.environ['OPENAI_API_KEY'] = "YOUR_OPENAI_API_KEY"

# Initialize the Large Language Model (LLM) that will power our agents
# Using GPT-4 for advanced reasoning and natural language understanding
llm = LLM(model="gpt-4")
```

```python
# ================================================================================
# STEP 1: Define User Inputs
# ================================================================================
# Define the user's fitness goals and preferences. This data will be used by
# our AI agents to create a personalized fitness plan.
# ================================================================================

fitness_goal = "muscle gain"
nutrition_preference = "high protein"

# Historical weight data helps agents understand progress trends
historical_weight_data = ["87kg", "85kg", "88kg", "87kg", "86kg"]

# ================================================================================
# STEP 2: Create AI Agents
# ================================================================================
# Agents are autonomous entities with specific roles and goals. Each agent has:
# - role: What the agent does
# - backstory: Context about the agent's expertise
# - goal: What the agent aims to achieve
# - tools: External capabilities (optional)
# ================================================================================

# Agent 1: Fitness Tracker - Creates personalized fitness plans
fitness_tracker_agent = Agent(
    llm=llm,
    role="Fitness Tracker",
    backstory="An AI-powered fitness tracker agent that helps users set goals, track progress, and recommend personalized workout plans.",
    goal="Set fitness goals, track user progress, and provide personalized recommendations based on user preferences.",
    # verbose=True,  # Uncomment to see detailed agent reasoning
)

# Agent 2: Recommendation Agent - Searches for fitness and nutrition information
recommendation_agent = Agent(
    llm=llm,
    role="Recommendation Agent",
    backstory="An AI agent that fetches personalized fitness and nutrition recommendations based on user goals and preferences.",
    goal="Search for fitness and nutrition information to provide recommendations that match the user's goals.",
    tools=[SerperDevTool()],  # SerperDev tool enables web search capabilities
    # verbose=True,  # Uncomment to see detailed agent reasoning
)

# ================================================================================
# STEP 3: Define Tasks
# ================================================================================
# Tasks are specific assignments given to agents. Each task has:
# - description: What needs to be done
# - expected_output: What the task should produce
# - agent: Which agent is responsible for the task
# ================================================================================

# Task 1: Set fitness goals and track progress
set_and_track_goals = Task(
    description=f"Set the user's fitness goal to {fitness_goal} and track their progress with a "
                f"personalized plan, also take into account the last weight measurements: {historical_weight_data}.",
    expected_output="A fitness plan for muscle gain and progress tracking setup.",
    agent=fitness_tracker_agent,
)

# Task 2: Fetch personalized fitness and nutrition recommendations
fetch_fitness_recommendations = Task(
    description=f"Fetch fitness recommendations for goal {fitness_goal} and nutrition preference {nutrition_preference}.",
    expected_output="Personalized fitness and nutrition recommendations based on the user's goal and dietary preference.",
    agent=recommendation_agent,
)

# Task 3: Provide step-by-step workout plan
provide_workout_plan = Task(
    description="Provide a step-by-step workout plan to help the user achieve their fitness goal.",
    expected_output="A personalized workout plan to meet the muscle gain goal.",
    agent=fitness_tracker_agent,
)

# ================================================================================
# STEP 4: Create and Configure the Crew
# ================================================================================
# A Crew is a team of agents working together to accomplish tasks.
# - agents: List of agents in the crew
# - tasks: List of tasks to be completed
# - planning: If True, the crew creates an execution plan before starting
# - verbose: If True, shows detailed execution logs
# ================================================================================

crew = Crew(
    agents=[fitness_tracker_agent, recommendation_agent],
    tasks=[set_and_track_goals, fetch_fitness_recommendations, provide_workout_plan],
    planning=True,  # Enable intelligent task planning and coordination
    verbose=True,   # Show detailed execution progress
)
```

```python
# ================================================================================
# STEP 5: Execute the Crew
# ================================================================================
# The kickoff() method starts the crew execution. The crew will:
# 1. Create an execution plan (because planning=True)
# 2. Assign tasks to appropriate agents
# 3. Execute tasks in optimal order
# 4. Allow agents to collaborate and share information
# 5. Return the final result
# ================================================================================

print("ðŸš€ Starting Fitness Progress Tracker Crew...\n")
result = crew.kickoff()
print("\nâœ… Crew execution completed!")
```

```python
# ================================================================================
# STEP 6: Display Results in a Readable Format
# ================================================================================
# The result object contains the crew's output. We'll display it as formatted
# markdown for better readability in the notebook.
# 
# result.raw - Contains the raw text output from the final task
# result.tasks_output - Contains output from each individual task
# ================================================================================

from IPython.display import display, Markdown

# Display the final result as formatted markdown
display(Markdown(result.raw))

# Optional: Access individual task outputs
# for i, task_output in enumerate(result.tasks_output, 1):
#     print(f"\n{'='*60}")
#     print(f"Task {i} Output:")
#     print(f"{'='*60}")
#     print(task_output.raw)
```
